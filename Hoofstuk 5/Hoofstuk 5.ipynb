```python
#Jeandre Mostert
#24918717
#Chapter 5 Lab Assignment

import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets
def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'],columns = ds['feature_names'])
    code_species_map = dict(zip(range(3),ds['target_names']))
    df['species'] = [code_species_map[c] for c in ds['target']]
    return df
df = get_iris_df()

#All plots generated are saved as .jpg files, thus run the script once to generate all the diagrams.
#These diagram names are self-explanitory to easily find.

#Pie Charts
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='pie',fontsize=20)
plt.ylabel(var,horizontalalignment='left')
plt.title('Breakdown for '+var, fontsize=25)
plt.savefig('iris_pie_for_one_variable.jpg')
plt.close()
#Note a simple representation of a single pie chart with 3 elements

sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind='pie', subplots=True,layout=(2,2),legend=False)
plt.title('Total Measurements, by Species')
plt.savefig('iris_pie_for_each_variable.jpg')
plt.close()
#Note multiple pie charts comparing different columns of the data set

#Bar Charts
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='bar', fontsize=15, rot=30)
plt.title('breakdown for '+var,fontsize=20)
plt.savefig('iris_bar_for_one_variable.jpg')
plt.close()
#Note that a bar chart is a much better representation to compare exact values

sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind='bar', subplots=True, fontsize=12, rot=30)
plt.suptitle('Total Measurements, by Species')
plt.savefig('iris_bar_for_each_variable.jpg')
plt.close()
#Note that each column is compared in 4 different bar charts, this shows how large the difference in size is.

#Histograms
df.plot(kind='hist', subplots=True, layout=(2,2))
plt.suptitle('Iris Histograms', fontsize=20)
#plt.show()
plt.close()

for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(kind='hist',alpha=0.4,label=spec)
plt.legend(loc='upper right')
plt.suptitle('Petal Length by Species')
plt.savefig('iris_hist_by_spec.jpg')
plt.close()
#Note that each species has its own histogram and color, but are plotted on the same axes to show the large difference in size

#Boxplots
col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i: i% 50)
df.pivot('ind','species')[col].plot(kind='box')
#plt.show()
plt.savefig('iris_boxplot_by_species.jpg')
plt.close()
#Note the wide distribution of data of the virginica plant species, displaying an outlayer and very wide 25/75% marks.

#Scatterplots
df.plot(kind="scatter",x="sepal length (cm)", y="sepal width (cm)")
plt.title("Length vs Width")
#plt.show()
plt.close()
colors = ["r", "g", "b"]
markers= [".", "*", "^"]
fig, ax = plt.subplots(1,1)
for i, spec in enumerate(df['species'].unique() ):
    ddf = df[df['species']==spec]
    ddf.plot(kind="scatter",x="sepal width (cm)", y="sepal length (cm)", 
             alpha =0.5, s=10*(i+1), ax=ax, color=colors[i], marker=markers[i],
            label = spec)
plt.legend()
#plt.show()
plt.savefig('iris_scatterplot_by_species.jpg')
plt.close()
#Note the distribution of the data, darker dots are overlapping points.

#Heatmaps
df.plot(kind="hexbin", x="sepal width (cm)", y="sepal length (cm)")
#plt.show()
plt.savefig('iris_heatmap.jpg')
plt.close()
#Note the color gradient indicating the density of datapoints.

#Time Series
import statsmodels.api as sm
dta = sm.datasets.co2.load_pandas().data
dta.plot()
plt.title("CO2 Levels")
plt.ylabel("Parts per million")
#plt.show()
plt.savefig("CO2_Time_Series.jpg")
plt.close()
#Note how with time the CO2 levels have been constantly increasing even with intermediate decreases.
```
